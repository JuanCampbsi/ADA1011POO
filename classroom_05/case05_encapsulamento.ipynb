```python
#TODO: Reimplementação de métodos

class Fraction:
    
    __numerator = None
    __denominator = None
    __cliente = [1,2,3,4,5,6]
    
    def __init__(self, numerator:int, denominator:int) -> None:
        """_summary_

        Args:
            numerator (int): _description_
            denominator (int): _description_
        """
        self.numerator = numerator
        self.denominator = denominator

    @property        
    def denominator(self):
        return self.__denominator
    
    @denominator.setter
    def denominator(self, denominator):
        if denominator == 0:
            raise ValueError('Erro! denominator não pode ser zero.')
        
        self.__denominator = denominator
        
    @property
    def numerator(self):
        return self.__numerator
    
    @numerator.setter
    def numerator(self, numerator):
        self.__numerator = numerator
        
    def __getitem__(self, key):
        return self.__cliente[key]
                
    def __add__(self, fraction):
        """_summary_

        Args:
            fraction (fraction): _description_

        Returns:
            fraction: _description_
        """
        numerator =  fraction.denominator * self.numerator + fraction.numerator * self.denominator
        denominator = fraction.denominator * self.denominator        
        return Fraction(numerator, denominator)
    
    def __sub__(self, fraction):        
        numerator =  fraction.denominator * self.numerator - fraction.numerator * self.denominator
        denominator = fraction.denominator * self.denominator        
        return Fraction(numerator, denominator)
    
    def __mul__(self, fraction):        
        numerator =  self.numerator * fraction.numerator
        denominator = fraction.denominator * self.denominator        
        return Fraction(numerator, denominator)
    
    def __truediv__(self, fraction):        
        numerator =  self.numerator * fraction.denominator
        denominator = fraction.numerator * self.denominator        
        return Fraction(numerator, denominator)
    
    @property
    def preview(self):
        return f"{self.numerator}/{self.denominator}"
    
    def __gt__(self, other):
        return self.__numerator > other.__denominator
    
    def __ge__(self, other):
        return self.__numerator >= other.__denominator
    
    def __lt__(self, other):
        return self.__numerator < other.__denominator
    
    def __le__(self, other):
        return self.__numerator <= other.__denominator
    
    def __eq__(self, other):
        return self.__numerator == other.__denominator
    
    def __ne__(self, other):
        return self.__numerator != other.__denominator

    
f1 = Fraction(5,5)
f2 = Fraction(3,4)

resultado_sum = f1 + f2
resultado_sub = f1 - f2
resultado_mult = f1 * f2
resultado_div = f1 / f2
resultado_get_item = f1[5]

print("\n----Comparações-----\n")
print("gt:", f1 > f2)
print("ge:", f1 >= f2)
print("lt:", f1 < f2)
print("le:", f1 <= f2)
print("eq:", f1 == f2)
print("ne:", f1 != f2)


print("\n----Operações-----\n")
print("soma:", resultado_sum.preview)
print("sub:", resultado_sub.preview)
print("mult:", resultado_mult.preview)
print("div:", resultado_div.preview)
print("getItem:", resultado_get_item)
```
